```python
import pandas as pd
import numpy as np
import utils
```

```python
# Get datasets
data = pd.read_csv("Original/train.csv", sep=';')
X_test = pd.read_csv("Original/test.csv", sep=';')
y_test = pd.read_csv("Original/test_labels.csv", sep=';')
hidden_data = pd.read_csv("Original/hidden_test.csv", sep=';')
```

```python
# encode categorical data
for feature in ["GROUP", "FORMATION", "WELL"]:
    if feature in data.columns:
        unique_list = data[feature].unique().tolist()
        feature_enc = {x:unique_list.index(x)+1 for x in unique_list}
        
        data[feature] = data[feature].map(feature_enc)
        
        
        for df in [X_test, hidden_data]:
            updated_list = unique_list
            df_list = df[feature].unique().tolist()

            for i in df_list:
                if i not in updated_list:
                    updated_list.append(i)
                    
            enc = {x:updated_list.index(x)+1 for x in updated_list}
            df[feature] = df[feature].map(enc)
```

```python
# Get standardising values to ensure same values are applied to all datasets
    
for feature in utils.Numerical:
    mean = data[feature].mean()
    std = data[feature].std()
    
    for d in [data,X_test,hidden_data]:
        d[feature] =  (d[feature] - mean)/std
        
```

```python
# Any cols with no data at all set to 0, i.e mean of training data
#data = data.fillna(0)
#X_test = X_test.fillna(0)
#hidden_data = hidden_data.fillna(0)
```

```python
# All missing data is set to 0. It represents dropout.
data = data.fillna(0)
X_test = X_test.fillna(0)
hidden_data = hidden_data.fillna(0)
```

```python
features = utils.Numerical + ["GROUP", "FORMATION", "WELL"]
```

```python
# Transform labels 

data[utils.y_name] = data[utils.y_name].map(utils.lithology_numbers)
y_test[utils.y_name] = y_test[utils.y_name].map(utils.lithology_numbers)
hidden_data[utils.y_name] = hidden_data[utils.y_name].map(utils.lithology_numbers)
```

```python
# Push to CSV

data[features].to_csv("X_train_0_WELL.csv", index = False)
data[utils.y_name].to_csv("Y_train.csv", index = False)

X_test[features].to_csv("X_test_0_WELL.csv", index = False)
y_test[utils.y_name].to_csv("Y_test.csv", index = False)

hidden_data[features].to_csv("X_hidden_0_WELL.csv", index = False)
hidden_data[utils.y_name].to_csv("Y_hidden.csv", index = False)
```
